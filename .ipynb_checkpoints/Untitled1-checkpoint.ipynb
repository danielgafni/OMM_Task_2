```python
from math import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import imageio

#steps number
Nx = 30
Ny = 30
M = 1000
#range
x0 = 0.0
x1 = 1.0
y0 = 0.0
y1 = 1.0
t0 = 0.0
t1 = 1.0
#step
hx = (x1 - x0) / Nx
hy = (y1 - y0) / Ny
ht = (t1 - t0) / M

#irregularity
def f(j, k, l):
	return cos(pi*k*hy) * exp(-l*ht)

#arrays
alphax = [[0.0 for j in range(Nx)] for k in range(Ny)]
betax = [[0.0 for j in range(Nx)] for k in range(Ny)]
pointx = [[0.0 for j in range(Nx)] for k in range(Ny)]
fx = [[0.0 for j in range(Nx)] for k in range(Ny)]
alphay = [[0.0 for j in range(Nx)] for k in range(Ny)]
betay = [[0.0 for j in range(Nx)] for k in range(Ny)]
pointy = [[0.0 for j in range(Nx)] for k in range(Ny)]
fy = [[0.0 for j in range(Nx)] for k in range(Ny)]

filenames = []
for l in range(M):
	########## находим pointx - значения на промежуточном слое
	
	ax = ht / (2.0 * hx**2)
	bx = ht / (2.0 * hx**2)
	cx = 1.0 + ht / hx**2
	
	#прямая прогонка
	
	#правый столбец
	for j in range(Nx):
		for k in range(Ny):
			km1 = k-1 if k-1 >= 0 else k
			kp1 = k+1 if k+1 < Ny else k
			if k == 0 or k == Ny - 1:
				fx[j][k] = 0.5*ht*f(j,k,l)
			else:
				fx[j][k] = 0.5 * ht / (hy**2) * (pointy[j][km1]+pointy[j][kp1]) + (1.0 - ht/(hy**2)) * pointy[j][k] + 0.5*ht*f(j,k,l)
	#левая часть
	for k in range(Ny):
		alphax[0][k] = 0.0
		betax[0][k] = 0.0
		for j in range(Nx-1):
			alphax[j+1][k]=bx/(cx-alphax[j][k]*ax)
			betax[j+1][k]=(ax*betax[j][k]+fx[j][k])/(cx-alphax[j][k]*ax)
	
	#обратная прогонка
	
	for k in range(Ny):
		pointx[Nx-1][k] = betax[Nx-1][k]/(1.0-alphax[Nx-1][k])
		for j in range(Nx)[Nx-1:0:-1]:
			pointx[j-1][k] = alphax[j][k]*pointx[j][k]+betax[j][k]
	
	#граничные условия
	for k in range(Ny):
		pointx[0][k] = 0.0
		pointx[Nx-1][k] = 0.0
	for j in range(Nx):
		pointx[j][0] = pointx[j][1]
		pointx[j][Nx-1] = pointx[j][Nx-2]
	
	######### находим pointy - значения на слое
	
	ay = ht / (2.0 * hy**2)
	by = ht / (2.0 * hy**2)
	cy = 1.0 + ht / hy**2
	
	#прямая прогонка
	
	#правый столбец
	for k in range(Ny):
		for j in range(Nx):
			jm1 = j-1 if j-1 >= 0 else j
			jp1 = j+1 if j+1 < Nx else j
			#if k == 0 or k == Ny - 1:
			#	fy[j][k] = 0.5*ht*f(j,k,l)
			#else:
			fy[j][k] = 0.5 * ht / (hx**2) * (pointx[jm1][k]+pointx[jp1][k]) + (1.0 - ht/(hx**2)) * pointx[j][k] + 0.5*ht*f(j,k,l)
	#левая часть
	for j in range(Nx):
		alphay[j][0] = 1.0
		betay[j][0] = 0.0
		for k in range(Ny-1):
			alphay[j][k+1]=by/(cy-alphay[j][k]*ay)
			betay[j][k+1]=(ay*betay[j][k]+fy[j][k])/(cy-alphay[j][k]*ay)
	
	#обратная прогонка
	
	for j in range(Nx):
		pointy[j][Ny-1] = betay[j][Ny-1]/(1.0-alphay[j][Ny-1])
		for k in range(Ny)[Ny-1:0:-1]:
			pointy[j][k-1] = alphay[j][k]*pointy[j][k]+betay[j][k]
	
	#граничные условия
	for k in range(Ny):
		pointy[0][k] = 0.0
		pointy[Nx-1][k] = 0.0
	for j in range(Nx):
		pointy[j][0] = pointy[j][1]
		pointy[j][Nx-1] = pointy[j][Nx-2]

	# plot
	if l%20 == 0:
		xn = np.arange(x0,x1,hx)
		ym = np.arange(y0,y1,hy)
		X, Y = np.meshgrid(xn, ym)
		fig = plt.figure()
		ax = fig.add_subplot(111, projection='3d')
		ax.set_zlim3d(-0.1, 0.01)
		surf = ax.plot_surface(X, Y, pointy, cmap='YlOrBr')
		plt.title('Solution')
		plt.xlabel('y')
		plt.ylabel('x')
		#plt.show()
		fname = 'frame' + str(l) + '.png'
		filenames += [fname]
		plt.savefig(fname, bbox_inches='tight')
		plt.close()

images = []
for filename in filenames:
	images.append(imageio.imread(filename))
imageio.mimsave('movie.gif', images)

```
